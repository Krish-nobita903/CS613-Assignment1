```python
from datasets import Features, Value

def retain_relevant_columns(dataset):
    # Define the desired columns
    columns_to_keep = ['id', 'text']

    # Remove columns not in the desired list
    return dataset.remove_columns([col for col in dataset.column_names if col not in columns_to_keep])

# Apply to each dataset
konkani_madlad_noisy_cleaned = retain_relevant_columns(konkani_madlad['noisy'])
konkani_madlad_clean_cleaned = retain_relevant_columns(konkani_madlad['clean'])
konkani_oscar_cleaned = retain_relevant_columns(konkani_oscar['train'])
konkani_sanghraha_cleaned = retain_relevant_columns(konkani_sanghraha1['train'])
konkani_culturax_cleaned = retain_relevant_columns(konkani_culturax['train'])

def add_index_as_id(dataset):
    # Create a new column 'id' based on the index of each row
    def add_id(example, idx):
        return {'id': str(idx), 'text': example['text']}

    # Add 'id' to each row and return a new dataset
    return dataset.map(lambda example, idx: add_id(example, idx), with_indices=True)

def standardize_and_add_id(dataset, text_field='text'):
    # Add index as ID if not already present
    if 'id' not in dataset.features:
        dataset = add_index_as_id(dataset)

    # Standardize the text field
    standardized = dataset.map(lambda x: {text_field: str(x[text_field])})

    # Define common features schema
    common_features = Features({
        'id': Value(dtype='string', id=None),
        'text': Value(dtype='string', id=None)
    })

    # Cast to common features schema and remove other columns
    standardized = standardized.cast(common_features)

    # Remove columns that are not in the schema
    columns_to_keep = set(common_features.keys())
    standardized = standardized.map(lambda x: {key: x[key] for key in columns_to_keep if key in x})

    return standardized

# Apply the function to each dataset
konkani_madlad_noisy_standardized = standardize_and_add_id(konkani_madlad_noisy_cleaned)
konkani_madlad_clean_standardized = standardize_and_add_id(konkani_madlad_clean_cleaned)
konkani_oscar_standardized = standardize_and_add_id(konkani_oscar_cleaned)
konkani_sanghraha_standardized = standardize_and_add_id(konkani_sanghraha_cleaned)
konkani_culturax_standardized = standardize_and_add_id(konkani_culturax_cleaned)
```

```python
from datasets import concatenate_datasets

combined_dataset = concatenate_datasets([
    konkani_madlad_noisy_standardized,
    konkani_madlad_clean_standardized,
    konkani_oscar_standardized,
    konkani_sanghraha_standardized,
    konkani_culturax_standardized
])

# Verify the combined dataset
print(combined_dataset)
```

```python
# Save dataset as CSV files
combined_dataset.to_csv('combined_dataset.csv')
```
